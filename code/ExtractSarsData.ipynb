```python
import pandas as pd
import sys, os
import numpy as np
import re

def read_data():
    path = "../data/sars"
    data = {}
    for filename in os.listdir(path):
        m, d, y = filename.split('.')[0].split('_')
        date = y + '-' + m + '-' + d
        d = pd.read_csv(path+'/'+filename)
        countryCum = list(d.iloc[:-1,0:2].set_index('Country').to_dict().values())[0]
        for c in countryCum:
            label = ''.join(filter(str.isalpha, c))
            value = re.search(r"\d+", str(countryCum[c]))
            if value:
                value = value.group(0)
            if 'HongKong' in label:
                label = 'HongKong'
            if 'Taiwan' in label:
                label = 'Taiwan'
            if 'Macao' in label:
                label = 'Macau'
            if 'Russia' in label:
                label = 'Russia'
            if 'Korea' in label:
                label = 'SouthKorea'
            if 'VietNam' in label:
                label = 'Vietnam'
            if 'ChinaGuangdongProvince' not in label:
                if label not in data:
                    data[label] = {date:float(value)} if value else {}
                else:
                    data[label][date] = float(value) if value else 0
    return data
```

```python
d = read_data()
df = pd.DataFrame(d)
```

```python
df = df.fillna(0)
```

```python
import datetime
def format_date(s):
    y, m, d = s.split('-')
    if len(m) < 2: m = '0' + m
    if len(d) < 2: d = '0' + d
    return y + '-' + m + '-' + d

date = date.values
```

```python
dateObj = list()
for d in date:
    x = d.split('-')
    dateObj.append(datetime.date(int(x[0]), int(x[1]), int(x[2])))
d = datetime.date(2003, 3, 17)
last = datetime.date(2003, 7, 12)
missed = list()
while d < last:
    if d not in dateObj:
        missed.append(d)
    d += datetime.timedelta(days=1)
missed
```

```python
new_date = list()
for d in date:
    x = d.split('-')
    new_date.append(datetime.date(int(x[0]), int(x[1]), int(x[2])).strftime('%Y-%m-%d'))
# data = np.append(np.expand_dims(np.asarray(new_date), axis=1), data, 1)
new_date
```

```python
for d in missed:
    a = np.empty((data.shape[1],))
    a[:] = np.nan
    data = np.vstack((data, a))
```

```python
new_index = np.concatenate((new_date, [m.strftime('%Y-%m-%d') for m in missed]))
new_df = pd.DataFrame(data, columns=df.columns, index=new_index)
```

```python
new_df = new_df.sort_index()
new_df = new_df.interpolate()
```
