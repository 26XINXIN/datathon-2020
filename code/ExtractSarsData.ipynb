```python
import pandas as pd
import sys, os
import numpy as np
import re

def read_data():
    path = "../data/sars"
    data = {}
    for filename in os.listdir(path):
        m, d, y = filename.split('.')[0].split('_')
        date = y + '-' + m + '-' + d
        d = pd.read_csv(path+'/'+filename)
        countryCum = list(d.iloc[:-1,0:2].set_index('Country').to_dict().values())[0]
        for c in countryCum:
            label = ''.join(filter(str.isalpha, c))
            value = re.search(r"\d+", str(countryCum[c]))
            if value:
                value = value.group(0)
            if 'HongKong' in label:
                label = 'HongKong'
            if 'Taiwan' in label:
                label = 'Taiwan'
            if 'Macao' in label:
                label = 'Macau'
            if 'Russia' in label:
                label = 'Russia'
            if 'Korea' in label:
                label = 'SouthKorea'
            if 'VietNam' in label:
                label = 'Vietnam'
            if 'ChinaGuangdongProvince' not in label:
                if label not in data:
                    data[label] = {date:float(value)} if value else {}
                else:
                    data[label][date] = float(value) if value else {}
    return data
```

```python
d = read_data()
df = pd.DataFrame(d)
```

```python
df = df.fillna(0)
```
